```python
def region_name(row):
    sigungu = row.Sigungu
    if row.Sigungu == '세종특별자치시':
        sigungu = '세종'
    return sigungu
df.region_name = df.apply(region_name, axis=1)
df.region_name.unique()
```

```python
# Region 값을 통일해준다.

def region_name(row):
    region = row.Region
    if row.Region == '광주광역시':
        region = '광주'
    if row.Region == '대구광역시':
        region = '대구'
    if row.Region == '대전광역시':
        region = '대전'
    if row.Region == '부산광역시':
        region = '부산'
    if row.Region == '서울시':
        region = '서울'
    if row.Region == '울산광역시':
        region = '울산'
    if row.Region == '인천광역시':
        region = '인천'
    if row.Region == '전북':
        region = '전라북도'
    if row.Region == '세종특별자치시':
        region = '세종'
    return region

df['Region'] = df.apply(region_name, axis=1)
```

```python
df.groupby('Region').count()
```

#### Title에서 공백을 제거한 문자열을 storename 컬럼에 저장한다.

```python
storename = df.Title.str.replace(' ','').str.strip()
df.insert(1, 'storename', storename)
df.head()
```

#### 네이버 블로그에서 검색할 지역명을 search_area에 저장한다.

```python
def make_search_word(row):
    search_area = row.Sigungu
    if len(row.Sigungu) <= 2:   # 시군구가 2글자 이하일 경우 (중구, 남구, 북구 등) Region으로 검색한다. ex) 서울 나주곰탕
        search_area = row.Region
    else:                       # 시군구가 3글자이상일 경우에는 시군구+가게명으로 검색한다. ex) 종로 나주곰탕
        search_area = row.Sigungu[:-1]
    return search_area

search_area = df.apply(make_search_word, axis=1)
df.insert(4, 'search_area', search_area)
df.head()
```

### 4. 식당명이 중복되는 경우를 제거한다.

```python
# 먼저 인덱스를 생성한다.
df.reset_index(inplace=True)
df.head()
```

```python
gtitle = df.groupby('storename').count()

# 식당명이 2개 이상인 가게의 리스트를 저장한다.
multiple_stores = gtitle[gtitle.post_num>1].sort_values('post_num').index.to_list()
```

```python
# 식당명이 2개 이상인 가게는 66,373개였다.
len(multiple_stores)
```

```python
# 중복되는 가게명은 같은 검색결과가 중복되어 나타나니 제거해주어야 한다.
df[df.storename.isin(multiple_stores)].sort_values('Title')
```

```python
# post_num이 가장 큰 값 하나만 남겨두고 나머지는 제거하기 위해 먼저 인덱스를 추출한다.
idxs = df[df.storename.isin(multiple_stores)].groupby('storename', as_index=False).max()['index']
idxs
```

```python
# 식당명이 중복되는 가게는 따로 msdf에 저장한다.
msdf = df[df['index'].isin(idxs)].sort_values('Title')
msdf.head()
```

```python
msdf.shape
```

```python
len(df.storename)
```

```python
df[df.storename.isin(multiple_stores)].shape, df[~df.storename.isin(multiple_stores)].shape
```

```python
# 식당명이 중복된 가게를 df에서 모두 제거한다.
df_deleted = df[~df.storename.isin(multiple_stores)]
```

```python
df_result = pd.concat([df_deleted,msdf])
df_result.shape
```
