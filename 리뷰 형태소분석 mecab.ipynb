```python
from konlpy.tag import Mecab
import pandas as pd
import numpy as np
import re
import time

from sklearn.feature_extraction.text import CountVectorizer
```

```python
class review_anal:        
    
    # 리뷰(csv파일) 리스트로 가져오기
    def get_review(review):
    
        raw_review = pd.read_csv(review, encoding='utf-8')['Content']
    
        # 리뷰 텍스트 리스트에 담기
        raw_review_list=[]
        hangeul = re.compile('[^ ㄱ-ㅣ가-힣]+')
        for i in range(len(raw_review)):
            try:
                raw_review_list.append(hangeul.sub('',raw_review.loc[i]))
            except:
                pass
        return raw_review_list    
    
    
    # 리뷰 토큰화 전
    def token_review(
#                         ngram_range="(1,1)"
#                         , tokenizer=
                    ):
        # stop_words
        stop_words = set(['은','는','이','가','하','아','것','들','의','있','그','되','수','보','주','등','한'
                          ,'나','같','우리','때','년','지','오','말','일','그렇','위하','때문','그것'
                          ,'두','말하','알','그러나','받','일','또','그런','문제','그리고','크','중','나오','가지'
                          ,'시키','지금','생각하','그러','속','하나','집','살','적','월','데','자신','안','어떤'
                          ,'내','경우','명','생각','시간','그녀','다시','이런','앞','보이','번','나','다른','어떻','전','들'
                          ,'사실','이렇','점','싶','말','정도','좀','원','통하','소리','놓'
                          ,'error','으로','잇님','잇님들'])
    
        # BOW : 토큰-벡터
        mecab = Mecab()
        vect = CountVectorizer(
                                ngram_range=(2,3)
                                , tokenizer=mecab.morphs
                                , stop_words=stop_words
#                                 , max_df=   # 0.0~1.0 사이 실수(%) 혹은 정수(개수)
                                , min_df = 25  # 0.0~1.0 사이 실수(%) 혹은 정수(개수)
#                                 , max_feature=   # 추출할 피쳐의 개수 지정(정수)
                              )    
        return vect
    
    
    
    
    # fit, 어휘사전
    def vocabulary_review(review):
        vect = review_anal.token_review()
        vect.fit(review_anal.get_review(review))
        voca = vect.vocabulary_
        return voca
        
        
    # bag_of_words_list    
    def bow_review(review):
        vect = review_anal.token_review()
#         fitting = review_anal.vocabulary_review(review)
        vect.fit(review_anal.get_review(review))
        
        bag_of_words = vect.transform(review_anal.get_review(review))
        bag_of_words_list = bag_of_words.toarray()
        
        return bag_of_words_list
    
    
    
    # bag_of_words_names_list
    def names_review(review):
        vect = review_anal.token_review()
        vect.fit(review_anal.get_review(review))
        
        bag_of_words_names = vect.get_feature_names()
        
        return bag_of_words_names
```

```python
%%time

names_review_result= review_anal.names_review('./post_crawling_통합_정제_2.csv')

df = pd.DataFrame({
    'names' : names_review_result
})

df.to_csv('./post_crawling_통합_정제_2_names.csv', encoding='utf-8')
```
