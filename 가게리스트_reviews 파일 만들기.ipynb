### 2. 가게리스트 불러오기 - df_storelist

```python
df_storelist = pd.read_csv('./train/data/storelist.csv', encoding='utf-8')
df_storelist.drop('Unnamed: 0', axis=1, inplace=True)
df_storelist.head(2)
```

```python
df_storelist.shape
```

```python
df_storelist['reviews_num'] = df_storelist.storeid.map(storeid_review_count)
```

```python
df_storelist_w_review = df_storelist[df_storelist.reviews_num.notna()]
```

```python
df_storelist_w_review.shape
```

```python
df_storelist_w_review.reviews_num = df_storelist_w_review.reviews_num.astype(int)
df_storelist_w_review.head(3)
```

```python
df_storelist_w_review.set_index('storeid', inplace=True)
df_storelist_w_review.head(2)
```

```python
df_storelist_w_review.sort_values('reviews_num', ascending=False)
```

```python
df_storelist_w_review.reviews_num.describe()
```

```python
df_storelist_w_review.to_csv('./train/data/가게리스트_reviews.csv', encoding='ut')
```
