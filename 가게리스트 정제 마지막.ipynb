```python
from collections import defaultdict
random_dic = defaultdict(lambda x : 0)
```

```python
# 랜덤으로 부여할 카테고리의 이름을 random_dic 에 인덱스와 함께 저장한다.
korean_random_category = korean_random_category.index.to_list()
for i in range(len(korean_random_category)):
    random_dic[korean_random_category[i]] = i
```

```python
# random_dic의 key와 value값을 서로 바꾼다.
random_dic = dict(random_dic)
random_dic = {idx : cate for cate, idx in random_dic.items()}
```

```python
# Category2 가 '알수없음'인 행들의 index를 unk_idxs 리스트에 담는다.
unk_idxs = df[df.Category2 == '알수없음'].index
```

```python
# unk_idxs의 인덱스들에 각각 0~59까지의 랜덤 int 값을 부여한다.
unk_dic = defaultdict(lambda x: 0)
for idx in unk_idxs.to_list():
    unk_dic[idx] = np.random.randint(60)
```

```python
# unk index, category index, category 이름을 저장할 데이터프레임을 생성한다.
unk_df = pd.DataFrame({
    'unk_idx' : unk_idxs
})
unk_df.head()
```
